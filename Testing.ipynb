WEBCAM TESTING

```python
import cv2
from ultralytics import YOLO

model = YOLO("D:/PaperRecyclabilityDetection/runs/detect/train/weights/best.pt")

cap = cv2.VideoCapture(1)

while True:
    ret, frame = cap.read()  
    if not ret:
        break

    results = model.predict(frame,conf=.5)
    annotated_frame = results[0].plot()
    cv2.imshow("Webcam YOLOv8", annotated_frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        cap.release()  
        cv2.destroyAllWindows()  
        os._exit(0)  
```

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

model = YOLO("D:/PaperRecyclabilityDetection/runs/detect/train/weights/best.pt")

image_path = "D:/PaperRecyclabilityDetection/TestSet/CorrectSamples"
results = model.predict(source=image_path,save=True, conf=0.5, iou=0.5,name="CorrectOutput")

for result in results:
    annotated_image = result.plot()  

plt.imshow(annotated_image)
plt.axis('off')
plt.show()
```
